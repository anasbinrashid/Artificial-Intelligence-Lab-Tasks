### Task 1


```python
import random

class Environment:
    def __init__(self, size=2):
        self.size = size
        self.grid = [['Clean' for _ in range(size)] for _ in range(size)]
        self.dirty()

    def dirty(self):
        for i in range(self.size):
            for j in range(self.size):
                if random.random() < 0.5:
                    self.grid[i][j] = 'Dirty'

    def location(self, x, y):
        if 0 <= x < self.size and 0 <= y < self.size:
            return self.grid[x][y]
        else:
            return "Invalid Location"

class Agent:
    def __init__(self, environment):
        self.environment = environment
        self.x = random.randint(0, environment.size-1)
        self.y = random.randint(0, environment.size-1)

    def sense_environment(self):
        return self.environment.location(self.x, self.y)

env = Environment(5)
agent = Agent(env)
print("Current Position of Agent: (", agent.x, ",", agent.y, ")")
print("Current State of Agent's Position: ", agent.sense_environment())
print("Current State of Grid: ")

for row in env.grid:
    print(row)
```

### Task 2

```python
class SimpleReflexAgent(Agent):
    def __init__(self, environment):
        super().__init__(environment)

    def act(self):
      if self.sense_environment() == 'Dirty':
          if 0 <= self.x < self.environment.size and 0 <= self.y < self.environment.size:
              self.environment.grid[self.x][self.y] = 'Clean'
          print("Cleaned cell (", self.x, ",", self.y, ")")
      else:
          print(f"Cell ({self.x}, {self.y}) is already clean")

environment = Environment(4)
agent = SimpleReflexAgent(environment)

print("Initial Grid:")
for row in environment.grid:
    print(row)

check = True

for i in range(100):
    agent.act()
    agent.x = random.randint(0, environment.size-1)
    agent.y = random.randint(0, environment.size-1)

    for row in environment.grid:
        if 'Dirty' in row:
            check = False
            break

    if check == False:
        check = True
    else:
        print("Grid Cleaned, terminating in ", i, "iterations")
        break

print("Final Grid:")
for row in environment.grid:
    print(row)
```

### Task 3

```python
class ModelBasedReflexAgent(Agent):
    def __init__(self, environment):
        super().__init__(environment)
        self.environment = environment
        self.x = random.randint(0, environment.size - 1)
        self.y = random.randint(0, environment.size - 1)
        self.memory = {}
        self.steps = 0
        self.cleaned = 0

    def act(self):
        self.steps += 1
        self.memory[(self.x, self.y)] = self.sense_environment()
        if self.memory[(self.x, self.y)] == 'Dirty':
            self.environment.grid[self.x][self.y] = 'Clean'
            self.cleaned += 1
            print("Cleaned cell (", self.x, ",", self.y, ")")
            self.memory[(self.x, self.y)] = 'Clean'
        else:
            print(f"Cell ({self.x}, {self.y}) is already clean")

        self.move()

    def move(self):
        available = [
            (self.x + x1, self.y + x2) for x1, x2 in [(-1, 0), (1, 0), (0, -1), (0, 1)]
            if 0 <= self.x + x1 < self.environment.size and 0 <= self.y + x2 < self.environment.size
        ]

        freetogo = [pos for pos in available if pos not in self.memory or self.memory[pos] == 'Dirty']

        if freetogo:
            self.x, self.y = random.choice(freetogo)
        else:
            self.x, self.y = random.choice(available)

        print("Moving to (", self.x, ",", self.y, ")")

environment = Environment(4)
agent = ModelBasedReflexAgent(environment)

print("Initial Grid:")
for row in environment.grid:
    print(row)

check = True

for i in range(100):
    agent.act()

    for row in environment.grid:
        if 'Dirty' in row:
            check = False
            break

    if check == False:
        check = True
    else:
        print("Grid Cleaned, terminating in", i, "iterations")
        break

print("Final Grid:")
for row in environment.grid:
    print(row)
```

### Task 4

```python
class PerformanceAgent(SimpleReflexAgent, ModelBasedReflexAgent):
    def __init__(self, environment):
        super().__init__(environment)
        self.steps = 0
        self.cleaned = 0

    def act(self):
        self.steps += 1
        if self.sense_environment() == 'Dirty':
            self.environment.grid[self.x][self.y] = 'Clean'
            self.cleaned += 1
            print(f"Cleaned cell ({self.x}, {self.y})")
        else:
            print(f"Cell ({self.x}, {self.y}) is already clean")

        self.x = random.randint(0, self.environment.size - 1)
        self.y = random.randint(0, self.environment.size - 1)

def run_agent(agent_class, env_size=4, max_steps=50):
    environment = Environment(env_size)
    agent = agent_class(environment)

    print(f"\nRunning {agent_class.__name__} \nInitial Grid:")
    for row in environment.grid:
        print(row)

    for i in range(max_steps):
        agent.act()

        if all("Dirty" not in row for row in environment.grid):
            print(f"Grid Cleaned, terminating in {i + 1} iterations")
            break

    efficiency_score = (agent.cleaned / agent.steps) * 100 if agent.steps > 0 else 0

    print("Final Grid:")
    for row in environment.grid:
        print(row)

    print(f"\n{agent_class.__name__} Performance:")
    print(f"Total Steps Taken: {agent.steps}")
    print(f"Total Dirt Cleaned: {agent.cleaned}")
    print(f"Efficiency Score: {efficiency_score:.2f}%\n")
    return agent.steps, efficiency_score

simple_steps, simple_score = run_agent(PerformanceAgent)
model_steps, model_score = run_agent(ModelBasedReflexAgent)

print("\nPerformance Comparison:")
print(f"\nSimple Reflex Agent\nSteps: {simple_steps}, Efficiency: {simple_score:.2f}%")
print(f"\nModel-Based Reflex Agent\nSteps: {model_steps}, Efficiency: {model_score:.2f}%")

if model_score > simple_score:
    print("\nModel-Based Reflex Agent is more efficient!")
else:
    print("\nSimple Reflex Agent performed better or equally efficient.")
```
